# Uso de la clase "Bono": 

```python
from bonos import  Bono #Clase definida en el archivo "bonos.py"
```

### Notación para definir un bono y guardarlo en una variable (las tasa de interés y cupones deben ser expresadas en términos porcentuales y anuales)

```python
ejemplo = Bono(nominal=100, tasa=10, cupon=5, plazo=10, frecuencia=2)
```

### Si se imprime, el bono muestra sus características (las tasas son convertidas y el precio es calculado en automático)

```python
print(ejemplo)
```

## Duración y convexidad: 

### Se pueden calcular las duraciones y convexidades modificados y de Maculay, debe expresarse "mod" o "mac" respectivamente

```python
print(ejemplo.duracion("mod"))
print(ejemplo.duracion("mac"))
```

```python
print(ejemplo.convexidad("mod"))
print(ejemplo.convexidad("mac"))
```

### Si el método se deja en blanco, arroja por default la duración ó convexidad de Maculay (mac)

```python
print(ejemplo.duracion())
print(ejemplo.convexidad())
```

### También se pueden calcular los factores de convexidad y duración para periodos específicos en el tiempo

```python
print(ejemplo.factor_duracion(5))
print(ejemplo.factor_convexidad(5))
```

## Resúmen del bono \\
